```python
from glob import glob
from moviepy.editor import *
import os

CorpusDir = './Game_data/Super Metroid/expert/'
imageDir = CorpusDir + 'screenshots/'
images = glob(imageDir + '/*.png')
images = sorted(images, key=lambda x: int(os.path.basename(x)[:-4]))  
clips = [ImageClip(m).set_duration(1/24)
         for m in images]


```

```python
concat_clip = concatenate_videoclips(clips, method="compose")
concat_clip.write_videofile(CorpusDir + "/movie.mp4", fps=24)
```
